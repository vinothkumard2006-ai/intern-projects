```python
#category
def get_category(pl):
    if '_PC_' in pl:
        return 'PC'
    elif '_SUP_' in pl:
        return 'Supplies'
    elif '_PH_' in pl:
        return 'Print Hardware'
    else:
        return 'Others'

#sub category
def get_sub_category(pl):
    if 'Laptop' in pl:
        return 'Laptop'
    elif 'Monitor' in pl:
        return 'Monitor'
    elif 'Desktop' in pl:
        return 'Desktop'
    elif 'INK' in pl:
        return 'Ink'
    elif 'TON' in pl:
        return 'Toner'
    elif 'LJ' in pl:
        return 'Laserjet'
    elif 'IJ' in pl:
        return 'Inkjet'
    elif 'SJ' in pl:
        return 'Single function Inkjet'
    elif 'UNK' in pl:
        return 'Unknown'
    else:
        return 'Others'

PLtable['CATEGORY'] = PLtable['PL'].apply(get_category)
PLtable['SUB_CATEGORY'] = PLtable['PL'].apply(get_sub_category)
```

```python
PLtable['SUB_CATEGORY'].unique()
```

```python
PLtable = PLtable[-PLtable['SUB_CATEGORY'].isin(['Single function Inkjet','Unknown'])]
PLtable = PLtable.reset_index(drop=True)
```

```python
def get_brand(pl):
    if pl.startswith('HP'):
       return 'HP'
    elif pl.startswith('DELL'):
       return 'DELL'
    else:
        return 'Others'
Price_list_table['BRAND'] =Price_list_table['PL'].apply(get_brand)
```

```python
merge_table = Price_list_table.merge(PLtable[['sku','CATEGORY','SUB_CATEGORY']],
                                     on='sku',how='left')
merge_table
```

```python

lpp_percentage ={
    'Laptop' :{'HP':0.0340,'DELL':0.0780},
    'Monitor':{'HP':0.0120,'DELL':0.0560},
    'Toner':{'HP':0.0910,'DELL':0.0230},
    'Laserjet':{'HP':0.0650,'DELL':0.0350},
    'Inkjet':{'HP':0.0120,'DELL':0.0560},
    'Desktop':{'HP':0.0910,'DELL':0.0230},
    'Ink':{'HP':0.0650,'DELL':0.0350}
}

#Lpp logic column
def calculate_lpp(row):
    sub = row['SUB_CATEGORY']
    brand = row['BRAND']
    map_price = row['MAP']

    if sub in lpp_percentage and brand in lpp_percentage[sub]:
        return map_price*(1-lpp_percentage[sub][brand])
    else:
        return map_price

merge_table['LPP'] = merge_table.apply(calculate_lpp, axis=1)
```

```python
def assign_quarter(dt):
   month = dt.month

   if month in [11,12,1]:
       return 'Q1'       ## nov-jan
   elif month in [2,3,4]:
       return 'Q2'       ##feb-april
   elif month in [5,6,7]:
       return 'Q3'       ##may-july
   elif month in [8,9,10]:
       return 'Q4'       ##aug-oct

merge_table['Quarter'] = merge_table['datetime'].apply(assign_quarter)
```
